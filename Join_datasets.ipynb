```python
# combines our virtual data and real data together using pandas library 
import pandas as pd
# read FIFA players ratings in
fifa_data21 = pd.read_csv('Virtual_data/player_ratings21.csv') 
fifa_data20 = pd.read_csv('Virtual_data/player_ratings20.csv') 
fifa_data19 = pd.read_csv('Virtual_data/player_ratings19.csv') 
fifa_data18 = pd.read_csv('Virtual_data/player_ratings18.csv') 
fifa_data17 = pd.read_csv('Virtual_data/player_ratings17.csv') 
fifa_data16 = pd.read_csv('Virtual_data/player_ratings16.csv') 
fifa_data15 = pd.read_csv('Virtual_data/player_ratings15.csv') 
# read results data in
results_data = pd.read_csv('Real_data/results_data.csv') 
# splits the seasons down into individual seasons
def season_data(results_data,season):
    season_data = (results_data['Season'])
    seasons = results_data[(results_data['Season'] == season)]
    return seasons
# call seasons function
season_14_15 = season_data(results_data,'2014/15')
season_15_16 = season_data(results_data,'2015/16')
season_16_17 = season_data(results_data,'2016/17')
season_17_18 = season_data(results_data,'2017/18')
season_18_19 = season_data(results_data,'2018/19')
season_19_20 = season_data(results_data,'2019/20')
season_20_21 = season_data(results_data,'2020/21')

# use the pd merge function to join the season data together with virtual data
fifa_season_14_15 = pd.merge(season_14_15, fifa_data15, left_on='HomeTeam', right_on='Team', how='left')

fifa_season_15_16 = pd.merge(season_15_16, fifa_data16, left_on='HomeTeam', right_on='Team', how='left')

fifa_season_16_17 = pd.merge(season_16_17, fifa_data17, left_on='HomeTeam', right_on='Team', how='left')

fifa_season_17_18 = pd.merge(season_17_18, fifa_data18, left_on='HomeTeam', right_on='Team', how='left')

fifa_season_18_19 = pd.merge(season_18_19, fifa_data19, left_on='HomeTeam', right_on='Team', how='left')

fifa_season_19_20 = pd.merge(season_19_20, fifa_data20, left_on='HomeTeam', right_on='Team', how='left')

fifa_season_20_21 = pd.merge(season_20_21, fifa_data21, left_on='HomeTeam', right_on='Team', how='left')

final_dataset_home = pd.concat([fifa_season_14_15,fifa_season_15_16,fifa_season_16_17,fifa_season_17_18,fifa_season_18_19,fifa_season_19_20,fifa_season_20_21])

final_dataset_home = final_dataset_home.drop(['Team','Unnamed: 0', 'HTR'], axis=1)


final_dataset_home.rename(columns = {'Total_Rating':'HTR', 'team_potential':'HTP',
                                   'team_wage':'HTW','team_value':'HTV','Goalkeeper':' HGKR',
                               'Defence': 'HDefR','Midfield': 'HMidR','Attack': 'HAttR'}, inplace = True) 


final_dataset_home
```

```python
# use the pd merge function to join the season data together with virtual data

fifa_season_14_15 = pd.merge(season_14_15, fifa_data15, left_on='AwayTeam', right_on='Team', how='left')

fifa_season_15_16 = pd.merge(season_15_16, fifa_data16, left_on='AwayTeam', right_on='Team', how='left')

fifa_season_16_17 = pd.merge(season_16_17, fifa_data17, left_on='AwayTeam', right_on='Team', how='left')

fifa_season_17_18 = pd.merge(season_17_18, fifa_data18, left_on='AwayTeam', right_on='Team', how='left')

fifa_season_18_19 = pd.merge(season_18_19, fifa_data19, left_on='AwayTeam', right_on='Team', how='left')

fifa_season_19_20 = pd.merge(season_19_20, fifa_data20, left_on='AwayTeam', right_on='Team', how='left')

fifa_season_20_21 = pd.merge(season_20_21, fifa_data21, left_on='AwayTeam', right_on='Team', how='left')

final_dataset_away = pd.concat([fifa_season_14_15,fifa_season_15_16,fifa_season_16_17,fifa_season_17_18,fifa_season_18_19,fifa_season_19_20,fifa_season_20_21])


final_dataset_away.rename(columns = {'Total_Rating':'ATR', 'team_potential':'ATP',
                                   'team_wage':'ATW','team_value':'ATV','Goalkeeper': 'AGKR',
                               'Defence': 'ADefR','Midfield': 'AMidR','Attack': 'AAttR'}, inplace = True)

final_dataset_away = final_dataset_away.drop(['Team','Unnamed: 0'], axis=1)
final_dataset_away
```

```python
# join the away team data with the home team data
final_dataset_home['ATR'] = final_dataset_away['ATR'] 
final_dataset_home['AGKR'] = final_dataset_away['AGKR'] 
final_dataset_home['ADefR'] = final_dataset_away['ADefR'] 
final_dataset_home['AMidR'] = final_dataset_away['AMidR'] 
final_dataset_home['AAttR'] = final_dataset_away['AAttR'] 
final_dataset_home['ATV'] = final_dataset_away['ATV'] 
final_dataset_home['ATP'] = final_dataset_away['ATP'] 
final_dataset_home['ATW'] = final_dataset_away['ATW'] 

final_dataset_home
```
