```python
import sqlite3 as sql
import pandas as pd
import numpy as np


results_database = 'results_data.db'
connection = sql.connect(results_database)
```

```python
# FIFA15 and 2014/15 Season 
sql_query = '''

SELECT season2014_15.HomeTeam, season2014_15.AwayTeam, season2014_15.FTR ,season2014_15.FTHG, season2014_15.FTAG, season2014_15.HS, season2014_15.ASH,season2014_15.HST,season2014_15.AST, season2014_15.HF, season2014_15.AF, season2014_15.HC, season2014_15.AC, season2014_15.HY, season2014_15.AY, season2014_15.HR, season2014_15.AR, season2014_15.B365H, season2014_15.B365D, season2014_15.B365A ,player_ratings15.Total_Rating, player_ratings15.Team_Potential, player_ratings15.Team_Wage, player_ratings15.Team_Value
FROM season2014_15
INNER JOIN player_ratings15 ON season2014_15.HomeTeam=player_ratings15.Team;
'''
fifa_results_home = pd.read_sql_query(sql_query, connection)

sql_query = '''

SELECT season2014_15.HomeTeam, season2014_15.AwayTeam, season2014_15.FTR ,season2014_15.FTHG, season2014_15.FTAG, season2014_15.HS, season2014_15.ASH,season2014_15.HST,season2014_15.AST, season2014_15.HF, season2014_15.AF, season2014_15.HC, season2014_15.AC, season2014_15.HY, season2014_15.AY, season2014_15.HR, season2014_15.AR, season2014_15.B365H, season2014_15.B365D, season2014_15.B365A ,player_ratings15.Total_Rating, player_ratings15.Team_Potential, player_ratings15.Team_Wage, player_ratings15.Team_Value, player_ratings15.season  
FROM season2014_15
INNER JOIN player_ratings15 ON season2014_15.AwayTeam=player_ratings15.Team;
'''
fifa_results_away = pd.read_sql_query(sql_query, connection)

fifa_results_home.rename(columns = {'Total_Rating':'HTR', 'team_potential':'HTP',
                                   'team_wage':'HTW','team_value':'HTV'}, inplace = True) 

fifa_results_away.rename(columns = {'Total_Rating':'ATR', 'team_potential':'ATP',
                                   'team_wage':'ATW','team_value':'ATV'}, inplace = True) 

final_dataset_2014_15 = pd.merge(fifa_results_home, fifa_results_away)
final_dataset_2014_15
```
