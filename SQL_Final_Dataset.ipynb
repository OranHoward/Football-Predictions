```python
import sqlite3 as sql
import pandas as pd
import numpy as np
```

```python
results_database = 'results_data.db'
connection = sql.connect(results_database)
sql_query1 = '''SELECT * FROM results_data_final'''
sql_query2 = '''SELECT * FROM fifa_data'''
results_data_final = pd.read_sql_query(sql_query1,connection)
fifa_data_final = pd.read_sql_query(sql_query2,connection)
```

```python
# SQL INNER JOIN to join results table and FIFA table
# Joins all the home team rating, value, potential, and wages. 

sql_query = '''

SELECT results_data_final.FTR, results_data_final.ID, results_data_final.HomeTeam, results_data_final.AwayTeam, results_data_final.FTHG, results_data_final.FTAG,results_data_final.season, results_data_final.Hshot, results_data_final.Ashot,results_data_final.HST,results_data_final.AST, results_data_final.HF, results_data_final.AF, results_data_final.HC, results_data_final.AC, results_data_final.HY, results_data_final.AY, results_data_final.HR, results_data_final.AR, results_data_final.B365H, results_data_final.B365D, results_data_final.B365A ,fifa_data.Total_Rating, fifa_data.Team_Potential, fifa_data.Team_Wage, fifa_data.Team_Value  
FROM results_data_final
INNER JOIN fifa_data ON results_data_final.HomeTeam=fifa_data.Team;
'''
# call the SQL query 
fifa_results_home = pd.read_sql_query(sql_query, connection)
fifa_results_home

sql_query = '''
SELECT results_data_final.ID, fifa_data.Total_Rating, fifa_data.Team_Potential, fifa_data.Team_Wage, fifa_data.Team_Value  
FROM results_data_final
INNER JOIN fifa_data ON results_data_final.AwayTeam=fifa_data.Team;
'''
fifa_results_away = pd.read_sql_query(sql_query, connection)
fifa_results_home

fifa_results_home.rename(columns = {'Total_Rating':'HTR', 'Team_Potential':'HTP',
                                   'Team_Wage':'HTW','Team_Value':'HTV'}, inplace = True) 

fifa_results_away.rename(columns = {'Total_Rating':'ATR', 'Team_Potential':'ATP',
                                   'Team_Wage':'ATW','Team_Value':'ATV'}, inplace = True) 

fifa_results_away

final_dataset = pd.merge(fifa_results_home, fifa_results_away)

final_dataset.to_csv(r'/Users/oran/Desktop/fifa_results_final.csv')
final_dataset
```
